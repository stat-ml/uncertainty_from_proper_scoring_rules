```python
def classify_strategy(metric_name):
    if "Excess" in metric_name:
        if "Inner" in metric_name:
            if metric_name.index("Outer") < metric_name.index("Inner"):
                return "Outer"
            else:
                return "Inner"
        else:
            return "Outer"
    elif "Outer" in metric_name:
        return "Outer"
    elif "Inner" in metric_name:
        return "Inner"
    else:
        return "Other"  # Handle cases without clear classification if any


ood_data["ApproximationStrategy"] = ood_data["UQMetric"].apply(classify_strategy)
results = ood_data.groupby("ApproximationStrategy")["RocAucScore"].agg(
    ["mean", "std", "count"]
)

sns.barplot(x=results.index, y=results["mean"])
plt.title("Average ROC AUC Score by Approximation Strategy")
plt.xlabel("Approximation Strategy")
plt.ylabel("Average ROC AUC Score")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the data
df = ood_data.copy()
same_loss_base = df[df["LossFunction"] == df["base_rule"]]
diff_loss_base = df[df["LossFunction"] != df["base_rule"]]
```

```python
# Calculate summary statistics
summary_same = same_loss_base["RocAucScore"].describe()
summary_diff = diff_loss_base["RocAucScore"].describe()

# Print summary statistics in a table
summary_table = pd.DataFrame(
    {
        "Same LossFunction and base_rule": summary_same,
        "Different LossFunction and base_rule": summary_diff,
    }
)
```

```python
# Plot the distributions
plt.figure(figsize=(14, 7), dpi=150)

pretty_matplotlib_config(fontsize=50)
sns.set(font_scale=3)  # crazy big

values = df["RocAucScore"].values
bins_ = np.linspace(values.min(), values.max(), 50)

# Distribution plot for same values
plt.subplot(1, 2, 1)
sns.histplot(same_loss_base["RocAucScore"], kde=True, color="blue", bins=bins_)
plt.title("Matching case")
plt.xlabel("AUROC")

# Distribution plot for different values
plt.subplot(1, 2, 2)
sns.histplot(diff_loss_base["RocAucScore"], kde=True, color="red", bins=bins_)
plt.title("Not-matching case")
plt.xlabel("AUROC")

plt.tight_layout()
plt.savefig("imgs/auroc_match_nomatch_ood.pdf")
plt.show()

# summary_table
```

```python
# Get unique LossFunction values
loss_functions = df["LossFunction"].unique()

# Create a figure with subplots for each LossFunction
n = len(loss_functions)
fig, axes = plt.subplots(n, 2, figsize=(14, n * 7), sharex=True, sharey=True, dpi=150)
fig.suptitle("Loss function stratification")

pretty_matplotlib_config(fontsize=50)
sns.set(font_scale=3)  # crazy big

# Iterate through each LossFunction value
for i, loss_function in enumerate(loss_functions):
    same_loss_base = df[
        (df["LossFunction"] == loss_function) & (df["LossFunction"] == df["base_rule"])
    ]
    diff_loss_base = df[
        (df["LossFunction"] == loss_function) & (df["LossFunction"] != df["base_rule"])
    ]

    # Plot distributions
    sns.histplot(
        same_loss_base["RocAucScore"], kde=True, color="blue", ax=axes[i, 0], bins=bins_
    )
    axes[i, 0].set_title(f"{loss_function} (Same)")
    axes[i, 0].set_xlabel("AUROC")

    sns.histplot(
        diff_loss_base["RocAucScore"], kde=True, color="red", ax=axes[i, 1], bins=bins_
    )
    axes[i, 1].set_title(f"{loss_function} (Different)")
    axes[i, 1].set_xlabel("AUROC")

plt.tight_layout()
plt.savefig("imgs/auroc_lossfunction_stratified_match_nomatch.pdf")
plt.show()
```

```python
# Get unique LossFunction and base_rule values
loss_functions = df["LossFunction"].unique()
base_rules = df["base_rule"].unique()

# Create a figure with subplots for each combination of LossFunction and base_rule
fig, axes = plt.subplots(
    len(base_rules), len(loss_functions), figsize=(20, 15), sharex=True, sharey=True
)
fig.suptitle("RocAucScore Distribution by LossFunction and base_rule")

# Iterate through each combination of LossFunction and base_rule
for i, base_rule in enumerate(base_rules):
    for j, loss_function in enumerate(loss_functions):
        subset = df[
            (df["base_rule"] == base_rule) & (df["LossFunction"] == loss_function)
        ]
        same_subset = subset[subset["Same"]]
        diff_subset = subset[~subset["Same"]]

        if not same_subset.empty:
            sns.histplot(
                same_subset["RocAucScore"],
                kde=True,
                color="blue",
                ax=axes[i, j],
                label="Same",
                bins=bins_,
            )
        if not diff_subset.empty:
            sns.histplot(
                diff_subset["RocAucScore"],
                kde=True,
                color="red",
                ax=axes[i, j],
                label="Different",
                bins=bins_,
            )

        axes[i, j].set_title(f"{base_rule} - {loss_function}")
        axes[i, j].set_xlabel("RocAucScore")
        axes[i, j].set_ylabel("")
        axes[i, j].legend()

plt.tight_layout()
plt.show()
```

```python
data = ood_data.copy()

# Extract patterns from the UQMetric column
data["Bayes_Pattern"] = data["UQMetric"].str.extract(r"(Inner|Outer)$")
data["Excess_Pattern"] = data["UQMetric"].str.extract(
    r"(Outer Inner|Outer Outer|Inner Outer)$"
)

# Display the first few rows to confirm the extraction
print(data[["UQMetric", "Bayes_Pattern", "Excess_Pattern"]].head())


# Separate the data into Bayes and Excess risk
bayes_data = data[data["RiskType"] == "Bayes"]
excess_data = data[data["RiskType"] == "Excess"]

# Merge the data on common columns to compare Bayes and Excess risk scores directly
comparison_df = pd.merge(
    bayes_data[
        [
            "Dataset",
            "LossFunction",
            "architecture",
            "base_rule",
            "Bayes_Pattern",
            "RocAucScore",
        ]
    ],
    excess_data[
        [
            "Dataset",
            "LossFunction",
            "architecture",
            "base_rule",
            "Excess_Pattern",
            "RocAucScore",
        ]
    ],
    left_on=["Dataset", "LossFunction", "architecture", "base_rule", "Bayes_Pattern"],
    right_on=["Dataset", "LossFunction", "architecture", "base_rule", "Excess_Pattern"],
    suffixes=("_Bayes", "_Excess"),
)

# Display the first few rows to confirm merging
print(comparison_df.head())


# Define the function to create scatter plots
def plot_scatter(data, stratification, title_suffix):
    unique_patterns = data[stratification].dropna().unique()
    if len(unique_patterns) == 0:
        print(f"No data available for stratification: {stratification}")
        return

    num_patterns = len(unique_patterns)
    fig, axes = plt.subplots(num_patterns, 1, figsize=(15, num_patterns * 5))

    for i, pattern in enumerate(unique_patterns):
        subset = data[data[stratification] == pattern]
        ax = axes[i] if num_patterns > 1 else axes
        sns.scatterplot(
            ax=ax,
            x="RocAucScore_Bayes",
            y="RocAucScore_Excess",
            hue="LossFunction",
            data=subset,
        )
        ax.plot([0.5, 1], [0.5, 1], "k--", lw=2)  # Line y=x for reference
        ax.set_xlabel("ROC AUC Score (Bayes Risk)")
        ax.set_ylabel("ROC AUC Score (Excess Risk)")
        ax.set_title(
            f"Scatter Plot of ROC AUC Scores: Bayes Risk vs. Excess Risk ({pattern} - {title_suffix})"
        )
        ax.legend(title="LossFunction")

    plt.tight_layout()
    plt.show()
```

```python
# Plot scatter plots stratified by Bayes_Pattern
plot_scatter(comparison_df, "Bayes_Pattern", "Stratified by Bayes_Pattern")

# Plot scatter plots stratified by Excess_Pattern
plot_scatter(comparison_df, "Excess_Pattern", "Stratified by Excess_Pattern")
```
